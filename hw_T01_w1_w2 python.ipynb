````python
#10. Имеются два $n$–мерных вектора $x$ и $y$, которые задают координаты $n$ точек на плоскости (случайные целые числа). 
#Найти наиболее близкие друг другу точки.

import numpy as np

def euclidean_distance(point1, point2):
    # Функция для вычисления евклидова расстояния между двумя точками
    return np.sqrt(np.sum((point1 - point2)**2))

def find_closest_points(x, y):
    min_distance = float('inf')
    closest_points = None

    for point_x in x:
        for point_y in y:
            distance = euclidean_distance(point_x, point_y)
            if distance < min_distance:
                min_distance = distance
                closest_points = (point_x, point_y)

    return closest_points, min_distance

# Пример использования
n = 3  # Замените на нужную размерность
x = np.random.randint(0, 100, (n,))  # Генерируем случайные точки для x
y = np.random.randint(0, 100, (n,))  # Генерируем случайные точки для y

closest_points, min_distance = find_closest_points(x, y)

print("Наиболее близкие точки:", closest_points)
print("Минимальное расстояние:", min_distance)
````
